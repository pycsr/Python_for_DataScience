```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# How to set plotting style

```python
sns.set_style("darkgrid") # white, dark, whitegrid, darkgrid
```

# Get Some toy dataset from Seaborn

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("tips")
df.head(2)
```

# Facet Frid

```python
fg = sns.FacetGrid(df,row="sex",col="day")
fg.map(sns.scatterplot,"tip","total_bill")
```

```python
fg = sns.FacetGrid(df,row="sex",col="day",hue="time")
fg.map(sns.scatterplot,"tip","total_bill")
fg.add_legend()
```

# KDE Plot and its variations

```python
df.head(2)
```

## Univariate plot (for single feature)

```python
sns.kdeplot(x=df.total_bill)
```

```python
sns.kdeplot(x=df.total_bill,hue=df.sex)
```

```python
sns.kdeplot(x=df.total_bill,hue=df.day)
```

## Bi-variate plot

```python
sns.kdeplot(x=df.tip,y=df.total_bill)
```

```python
sns.kdeplot(x=df.tip,y=df.total_bill,hue=df.sex)
```

```python
sns.kdeplot(x=df.tip,y=df.total_bill,hue=df.sex,shade=True)
```

```python
sns.kdeplot(x=df.tip,y=df.total_bill,hue=df.sex,shade=True,cbar=True)
```

# Joint Plot and its variations

```python
df.head(2)
```

```python
sns.jointplot(x=df.tip,y=df.total_bill)
```

```python
sns.histplot(df.tip)
```

```python
sns.histplot(df.total_bill)
```

```python
sns.jointplot(x=df.tip,y=df.total_bill) #"scatter" | "kde" | "hist" | "hex" | "reg"
```

```python
sns.jointplot(x=df.tip,y=df.total_bill,kind="kde") #"scatter" | "kde" | "hist" | "hex" | "reg"
```

```python
sns.jointplot(x=df.tip,y=df.total_bill,kind="hist") #"scatter" | "kde" | "hist" | "hex" | "reg"
```

```python
sns.jointplot(x=df.tip,y=df.total_bill,kind="hex") #"scatter" | "kde" | "hist" | "hex" | "reg"
```

```python
sns.jointplot(x=df.tip,y=df.total_bill,kind="reg") #"scatter" | "kde" | "hist" | "hex" | "reg"
```

```python
sns.jointplot(x=df.tip,y=df.total_bill,kind="reg",color="r") #"scatter" | "kde" | "hist" | "hex" | "reg"
```

# Regression Plot and its variations

# Pair Grid and its variations

```python
df.head(2)
```

```python
pg = sns.PairGrid(df)
pg.map_diag(sns.histplot)
pg.map_lower(sns.regplot)
pg.map_upper(sns.kdeplot)
```

```python
pg = sns.PairGrid(df,hue="sex")
pg.map_diag(sns.histplot)
pg.map_lower(sns.regplot)
pg.map_upper(sns.kdeplot)
```

# Clustermap Plot

```python
dfp = df.pivot_table(index=df.smoker,columns=df.day,values=["total_bill","tip"])
dfp
```

```python
sns.clustermap(dfp)
```

```python
sns.clustermap(dfp,figsize=(5,5),annot=True)
```

```python
sns.clustermap(dfp,figsize=(5,5),annot=True,cmap="copper_r")
```
