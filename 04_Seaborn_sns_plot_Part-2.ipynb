# Joint Plot and its variations

```python
sns.jointplot(df.tip,df.total_bill)
```

```python
sns.histplot(df.tip)
```

```python
sns.histplot(y=df.total_bill)
```

```python
sns.jointplot(df.tip,df.total_bill,kind="kde") # "scatter" | "kde" | "hist" | "hex" | "reg" | "resid"
```

```python
sns.jointplot(df.tip,kind="kde") # "scatter" | "kde" | "hist" | "hex" | "reg" | "resid"
```

```python
sns.jointplot(df.tip,df.total_bill,kind="hist") # "scatter" | "kde" | "hist" | "hex" | "reg" | "resid"
```

```python
sns.jointplot(df.tip,df.total_bill,kind="hex") # "scatter" | "kde" | "hist" | "hex" | "reg" | "resid"
```

```python
sns.jointplot(df.tip,df.total_bill,kind="reg") # "scatter" | "kde" | "hist" | "hex" | "reg" | "resid"
```

```python
sns.jointplot(df.tip,df.total_bill,kind="resid") # "scatter" | "kde" | "hist" | "hex" | "reg" | "resid"
```

# Pair Grid and its variations

```python
df.head(2)
```

```python
pg = sns.PairGrid(df)
pg.map_diag(sns.histplot)
pg.map_lower(sns.regplot)
pg.map_upper(sns.scatterplot)
```

```python
pg = sns.PairGrid(df)
pg.map_diag(sns.histplot)
pg.map_lower(sns.kdeplot)
pg.map_upper(sns.scatterplot)
```

```python
pg = sns.PairGrid(df,hue="sex")
pg.map_diag(sns.histplot)
pg.map_lower(sns.kdeplot)
pg.map_upper(sns.scatterplot)
pg.add_legend()
```
