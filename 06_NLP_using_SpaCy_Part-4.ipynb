# n-grams (using nltk)
bi-grams
tri-grams

```python
from nltk import bigrams,trigrams,ngrams
```

```python
a = "This is another good video in NLP using SpaCy"
print(a)
```

```python
word_token = a.split(" ")
print(word_token)
```

## bi-gram

```python
bi = bigrams(word_token)
print(bi)
print(list(bi))
```

## tri-gram

```python
bi = trigrams(word_token)
print(bi)
print(list(bi))
```

## N-gram

```python
bi = ngrams(word_token,4)
print(bi)
print(list(bi))
```

```python
bi = ngrams(word_token,1)
print(bi)
print(list(bi))
```

```python
bi = ngrams(word_token,2)
print(bi)
print(list(bi))
```

```python
bi = ngrams(word_token,3)
print(bi)
print(list(bi))
```

```python
bi = ngrams(word_token,1000) # Kilo-gram
print(bi)
print(list(bi))
```
