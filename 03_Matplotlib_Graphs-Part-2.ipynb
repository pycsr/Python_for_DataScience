# Editing property of Plots

```python
plt.figure(figsize=(16,4))
plt.scatter(x,y1)
plt.show()
```

## Labelling and marking in scatter plot

```python
plt.figure(figsize=(16,6))
for idx in range(len(y1)):
    plt.scatter(x[idx],y1[idx],c="r",alpha=.2,s=.25*np.pi*y1[idx]**2)
    if y1[idx] > 60:
        plt.text(x[idx],y1[idx],y1[idx])
plt.show()
```

```python
plt.figure(figsize=(16,6))
for idx in range(len(y1)):
    if y1[idx] > 60:
        plt.scatter(x[idx],y1[idx],c="r",alpha=.2,s=.25*np.pi*y1[idx]**2,marker="*")
        plt.text(x[idx],y1[idx],y1[idx])
    else:
        plt.scatter(x[idx],y1[idx],c="b",alpha=.2,s=.25*np.pi*y1[idx]**2,marker="<")
        
plt.show()
```

## Draw Horizontal or vertical line in plot

```python
plt.figure(figsize=(16,4))
plt.plot(x,y1)
plt.axvline(x=20,color="r",linewidth=4)
plt.axhline(y=60,color="r",linewidth=4)
# plt.annotate(text="This is Threshold cross",xy=(20,60),xytext=(30,80))
plt.annotate('This is Horizontal Line', xy=(20, 60), xytext=(30, 80),arrowprops=dict(facecolor='black'))
plt.show()
```

## Subplot and resizing

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y1)
plt.subplot(2,2,3)
plt.plot(x,y1)
plt.subplot(224)
plt.plot(x,y1)
plt.show()
```

## Subplot with shared X- axis

```python
ax = plt.subplots(2,1)
ax
```

```python
fig,ax = plt.subplots(2,1,sharex=True)
ax[0].plot(x,y1)
ax[1].scatter(x,y2)
plt.show()
```

## Subplot with shared Y- axis and X- axis

```python
fig,ax = plt.subplots(2,2,sharex=True,sharey=True)
ax
```

```python
fig,ax = plt.subplots(2,2,sharex=True,sharey=True)
ax[0][0].plot(x,y1)
ax[0][1].plot(x,y1)
ax[1][0].plot(x,y1)
ax[1][1].plot(x,y1)
plt.show()
```

# Animated plot

```python
%matplotlib
import matplotlib.animation as animation
xs = []
ys = []
fig = plt.figure(figsize=(16,4))
ax = plt.subplot(111)
def animate(i,xs,ys): # i is nothing but inbuilt variable to inbuilt variable to increment frame number
    ys.append(np.random.randint(-20,30))
    xs.append(len(ys))
    
    xs = xs[-20:]
    ys = ys[-20:]
    ax.clear()
    ax.plot(xs,ys)
    ax.scatter(xs,ys)


ani = animation.FuncAnimation(fig,func=animate,fargs=(xs,ys),interval=50)
plt.show()    
    
    
```

## 3D plot

```python
%matplotlib
import matplotlib.animation as animation
xs = []
ys = []
fig = plt.figure(figsize=(16,4))
ax = plt.subplot(111,projection="3d")
def animate(i,xs,ys): # i is nothing but inbuilt variable to inbuilt variable to increment frame number
    ys.append(np.random.randint(-20,30))
    xs.append(len(ys))
    
    xs = np.array(xs[-20:])
    ys = np.array(ys[-20:])
    zs = xs**2+ys**2
    ax.clear()
    for idx in range(len(ys)):
        ax.scatter(xs[idx],ys[idx],zs[idx],color="r",s=.03*zs[idx]**2,alpha=.4)

ani = animation.FuncAnimation(fig,func=animate,fargs=(xs,ys),interval=50)
plt.show()    
    
    
```
