# Dropping columns/rows `where NaN is present`

```python
a = np.random.randint(1,50,40).reshape(10,4)
df = pd.DataFrame(a,columns=list("ABCD"))
df
```

```python
df.loc[4,"B"] = np.NaN
df.loc[9,"C"] = np.NaN
df
```

## Drop rows (default)

```python
df.dropna()
```

## Drop columns

```python
df.dropna(axis=1)
```

## Dropping columns/rows of user choice if column/feature is not very important

```python
df.drop(["B","C","D"],axis=1)
```

```python
df
```
