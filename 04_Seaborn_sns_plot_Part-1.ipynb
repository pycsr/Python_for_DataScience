# Important libraries to import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Get Some toy dataset from Seaborn

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("tips")
df.head(5)
```

# How to set plotting style


```python
sns.set_style("white") # white, dark, whitegrid, darkgrid
```

# Count plot

```python
df.head(2)
```

```python
df.sex.value_counts().plot(kind="bar")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Count")
```

```python
sns.countplot(df.sex)
plt.title("Gender Count")
```
