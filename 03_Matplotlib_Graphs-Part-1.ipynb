## Smooting line plot

```python
from scipy.interpolate import interp1d
plt.figure(figsize=(16,4))
plt.plot(x,y1)
x_new = np.linspace(x.min(),x.max(),300)
inter = interp1d(x,y1,kind="zero")
inter2 = interp1d(x,y1,kind="linear")
inter3 = interp1d(x,y1,kind="quadratic")
plt.plot(x_new,inter(x_new),"m--")
plt.plot(x_new,inter2(x_new),"b-.")
plt.plot(x_new,inter3(x_new),"k.")
plt.show()
```

```python
from scipy.interpolate import interp1d
plt.figure(figsize=(16,4))
plt.plot(x,y1)
x_new = np.linspace(x.min(),x.max(),300)
inter3 = interp1d(x,y1,kind="quadratic")
plt.plot(x_new,inter3(x_new),"k--")
plt.show()
```

## Bubble plot

```python
plt.figure(figsize=(16,4))
plt.scatter(x,y1,color="b",s=200)
plt.show()
```

```python
plt.figure(figsize=(16,4))
for idx in range(len(y1)):
    plt.scatter(x[idx],y1[idx],color="b",s=.08*np.pi*y1[idx]**2)
plt.show()
```

```python
plt.figure(figsize=(16,4))
for idx in range(len(y1)):
    plt.scatter(x[idx],y1[idx],color="b",s=.08*np.pi*y1[idx]**2,alpha=0.2)
plt.show()
```

## Stacked Bar plot

```python
plt.figure(figsize=(16,4))
plt.bar(x,y1)
plt.bar(x,y2)
plt.show()
```

## Stacked Bar plot with labels

```python
plt.figure(figsize=(16,4))
for idx in range(len(y1)):
    plt.bar(x[idx],y1[idx])
plt.show()
```

```python
plt.figure(figsize=(16,4))
for idx in range(len(y1)):
    plt.bar(x[idx],y1[idx],color="r")
    plt.text(x[idx],y1[idx]+.8,y1[idx])
plt.show()
```

```python
plt.figure(figsize=(16,4))
for idx in range(len(y1)):
    plt.bar(x[idx],y1[idx],color="r")
    plt.text(x[idx]-.25,y1[idx],y1[idx])
plt.show()
```

## Stem

```python
x = np.linspace(1,50,50,dtype="int8")
y1 = np.random.randint(1,100,50)
y2 = np.random.randint(1,100,50)
```

```python
plt.figure(figsize=(16,4))
plt.stem(x,y1)
plt.show()
```

## Lolipop

```python
plt.figure(figsize=(16,4))
for idx in range(len(y1)):
    plt.scatter(x[idx],y1[idx],s=.03*3.14*y1[idx]**2)
plt.stem(x,y1)
plt.show()
```

## Area plot

```python
plt.figure(figsize=(16,4))
plt.fill_between(x,y1)
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.fill_between(x,y1,color="r")
plt.show()
```

## Mixed plot

```python
plt.figure(figsize=(16,4))
plt.plot(x,y1)
plt.bar(x,y1,width=.2)
plt.stem(x,y1)
plt.scatter(x,y1,color="r",s=200)
```

## Contour Plot

```python
x = np.arange(-2,2,.025)
y = np.arange(-3,3,.025)
X,Y = np.meshgrid(x,y)
Z = np.sin(X) ** 10 + np.cos(10 + Y * X) * np.cos(X)
cp = plt.contour(x,y,Z)
plt.clabel(cp)
plt.colorbar()
plt.show()
```


## Image Plot (Can be used for HeatMap)

```python
plt.imshow(y2)
for X in range(10):
    for Y in range(5):
        plt.text(X,Y,y2[Y][X])
```

```python
plt.imshow(y2,cmap="RdYlGn")
for X in range(10):
    for Y in range(5):
        plt.text(X,Y,y2[Y][X])
```
